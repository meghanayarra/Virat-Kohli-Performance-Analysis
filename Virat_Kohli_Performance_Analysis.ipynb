```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("virat_kohli.csv")
print(data.head())
```

The above dataset contain matches played by Virat Kohli between 18-August-2008 and 22-January-2017

```python
print(data.isnull())
```

```python
print(data.isnull().sum())
```

Total Runs scored by Virat

```python
data["Runs"].sum()
```

Average Runs scored by Virat

```python
data["Runs"].mean()
```

In ODI, batting average between 35-38 is a good average. So virat's batting average is good.

```python
y=data["Runs"]
plt.plot(data.index.values,y)
plt.title("Runs scored by Virat Kohli")
plt.xlabel("Index")
plt.ylabel("Runs")
plt.show()
```

In many innings played by Virat Kohli he scored over 100 or close to it.

```python
data["Pos"]=data["Pos"].map({1.0: "Batting At 1", 2.0: "Batting At 2", 3.0: "Batting At 3",
                               4.0: "Batting At 4", 5.0:"Batting At 5", 6.0:"Batting At 6",
                               7.0: "batting At 7"})
#print(data["Pos"])
Pos = data["Pos"].value_counts()
label = Pos.index
count= Pos.values
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "red", "orange"]
fig,ax=plt.subplots()
ax.pie(count,labels=label,colors=colors,autopct='%1.1f%%')
ax.legend(label,loc="upper right")
plt.title('Number of Matches At Different Batting Positions')
plt.show()
```

In more than 68% innings played by Virat Kohli he batted in third position

```python
centuries = data[data["Runs"] >= 100]
fig, ax = plt.subplots()
colors=plt.viridis()
ax.bar(centuries["Inns"], centuries["Runs"],color=colors)
plt.title("Centuries By Virat Kohli in First Innings Vs. Second Innings")
plt.xlabel("Innings")
plt.ylabel("Runs")

plt.show()
```

So Virat Kohli scored most of the centuries while batting in second innings.

```python
dismissal = data["Dismissal"].value_counts()
label = dismissal.index
counts = dismissal.values
colors = ['gold', 'lightgreen', 'pink', 'blue', 'skyblue', 'red', 'orange']

fig, ax = plt.subplots()
ax.pie(counts, labels=label, colors=colors, autopct=lambda p: f'{int(p * sum(counts) / 100)}')
ax.axis("equal")
plt.legend(label, loc= "upper left")
plt.title('Dismissals of Virat Kohli')
plt.show()
```

So most of the time, Virat Kohli gets out by getting caught by the fielder or the keeper.

```python

fig, ax = plt.subplots()
colors=plt.viridis()
bar_width = 0.8
bars = ax.bar(data["Opposition"], data["Runs"], color=colors, width=bar_width)
plt.title("Most Runs Against Teams")
plt.xlabel("Opposition")
plt.ylabel("Runs")
plt.xticks(rotation=45, ha="right")
plt.show()
```

According to the above figure, Virat Kohli likes batting against Sri Lanka, New Zealand, West Indies, and South Africa.

```python
fig, ax = plt.subplots()
colors=plt.viridis()
bar_width = 0.8
bars = ax.bar(centuries["Opposition"], centuries["Runs"], color=colors, width=bar_width)
plt.title("Most Centuries Against Teams")
plt.xlabel("Opposition")
plt.ylabel("Runs")
plt.xticks(rotation=45, ha="right")
plt.show()
```

So, most of the centuries scored by Virat Kohli were against New Zealand.

```python
strike_rate = data.query("SR >= 120")
print(strike_rate)
```

```python
fig, ax = plt.subplots()
colors=plt.viridis()
bar_width = 0.8
bars = ax.bar(strike_rate["Inns"], strike_rate["SR"], color=colors, width=bar_width)
plt.title("Virat Kohli's High Strike Rates in First Innings Vs. Second Innings")
plt.xlabel("Inns")
plt.ylabel("Strike Rate")
plt.show()
```

So according to the above figure, Virat Kohli likes playing in the first innings compared to the second innings

```python
fig, ax = plt.subplots()
scatter = ax.scatter(data["Runs"], data["4s"], s=data["SR"], c=data["SR"])
plt.title("Relationship Between Runs Scored and Fours")
plt.xlabel("Runs")
plt.ylabel("Fours")
plt.show()
```

There is a linear relationship. It means that Virat Kohli likes playing fours. The more runs he scores in the innings, the more fours he plays.

```python
fig, ax = plt.subplots()
scatter = ax.scatter(data["Runs"], data["6s"], s=data["SR"],c=data["SR"])
plt.title("Relationship Between Runs Scored and Fours")
plt.xlabel("Runs")
plt.ylabel("Sixs")
plt.show()
```

There is no strong linear relationship here. It means Virat Kohli likes playing fours more than sixes.
